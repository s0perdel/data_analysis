```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data - student test scores
scores = [65, 72, 58, 85, 90, 78, 82, 76, 88, 92, 68, 74, 81, 79, 85, 
          91, 67, 73, 84, 89, 77, 83, 70, 86, 80, 75, 87, 93, 69, 71]

plt.figure(figsize=(15, 4))

# Different bin sizes
plt.subplot(1, 3, 1)
plt.hist(scores, bins=5, color='lightblue', edgecolor='black')
plt.title('Too Few Bins (5) - Loss of Detail')
plt.xlabel('Test Scores')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(scores, bins=10, color='lightgreen', edgecolor='black')
plt.title('Good Number of Bins (10)')
plt.xlabel('Test Scores')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(scores, bins=20, color='lightcoral', edgecolor='black')
plt.title('Too Many Bins (20) - Too Noisy')
plt.xlabel('Test Scores')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Different types of distributions
np.random.seed(42)

# Create different datasets
normal_data = np.random.normal(100, 15, 1000)  # Normal distribution
skewed_data = np.random.exponential(50, 1000)   # Right-skewed
bimodal_data = np.concatenate([
    np.random.normal(80, 10, 500),
    np.random.normal(120, 10, 500)
])  # Bimodal distribution

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.hist(normal_data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(skewed_data, bins=30, alpha=0.7, color='orange', edgecolor='black')
plt.title('Right-Skewed Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(bimodal_data, bins=30, alpha=0.7, color='lightgreen', edgecolor='black')
plt.title('Bimodal Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Example 1: Age distribution in a company
ages = np.random.normal(35, 8, 500)  # Mean age 35, std 8
ages = np.clip(ages, 22, 65)  # Clip to realistic age range

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(ages, bins=20, color='lightblue', edgecolor='black', alpha=0.7)
plt.title('Employee Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.axvline(ages.mean(), color='red', linestyle='--', label=f'Mean: {ages.mean():.1f}')
plt.legend()

# Example 2: Monthly sales data
sales = np.random.gamma(shape=2, scale=5000, size=300)  # Right-skewed like real sales

plt.subplot(2, 2, 2)
plt.hist(sales, bins=15, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Monthly Sales Distribution')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.axvline(sales.mean(), color='red', linestyle='--', label=f'Mean: ${sales.mean():.0f}')
plt.legend()

# Example 3: Website loading times
load_times = np.random.exponential(2.5, 1000)  # Exponential for wait times

plt.subplot(2, 2, 3)
plt.hist(load_times, bins=25, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title('Website Loading Times')
plt.xlabel('Loading Time (seconds)')
plt.ylabel('Frequency')
plt.axvline(load_times.mean(), color='red', linestyle='--', label=f'Mean: {load_times.mean():.1f}s')
plt.legend()

# Example 4: Customer purchase amounts
purchases = np.random.lognormal(3, 1, 1000)  # Log-normal for monetary values

plt.subplot(2, 2, 4)
plt.hist(purchases, bins=20, color='plum', edgecolor='black', alpha=0.7)
plt.title('Customer Purchase Amounts')
plt.xlabel('Purchase Amount ($)')
plt.ylabel('Frequency')
plt.axvline(purchases.mean(), color='red', linestyle='--', label=f'Mean: ${purchases.mean():.1f}')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Compare two groups
np.random.seed(42)

# Male and female heights (different distributions)
male_heights = np.random.normal(178, 7, 500)  # Mean 178cm
female_heights = np.random.normal(165, 6, 500)  # Mean 165cm

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(male_heights, bins=20, alpha=0.7, label='Male', color='blue')
plt.hist(female_heights, bins=20, alpha=0.7, label='Female', color='pink')
plt.title('Height Distribution by Gender')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.legend()

# Side-by-side histogram
plt.subplot(1, 2, 2)
plt.hist([male_heights, female_heights], bins=15, 
         label=['Male', 'Female'], color=['blue', 'pink'], alpha=0.7)
plt.title('Height Distribution (Side-by-Side)')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Exam scores analysis
exam_scores = np.random.normal(75, 12, 200)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
# Regular histogram
counts, bins, patches = plt.hist(exam_scores, bins=15, color='lightblue', 
                                edgecolor='black', alpha=0.7)
plt.title('Exam Scores Distribution')
plt.xlabel('Scores')
plt.ylabel('Frequency')

# Add some statistics
plt.axvline(exam_scores.mean(), color='red', linestyle='--', label=f'Mean: {exam_scores.mean():.1f}')
plt.axvline(np.median(exam_scores), color='green', linestyle='--', label=f'Median: {np.median(exam_scores):.1f}')
plt.legend()

plt.subplot(1, 2, 2)
# Cumulative histogram
plt.hist(exam_scores, bins=15, color='orange', edgecolor='black', 
         alpha=0.7, cumulative=True)
plt.title('Cumulative Exam Scores Distribution')
plt.xlabel('Scores')
plt.ylabel('Cumulative Frequency')
plt.axvline(exam_scores.mean(), color='red', linestyle='--', label=f'Mean: {exam_scores.mean():.1f}')

# Add percentage lines
for percentile in [25, 50, 75]:
    value = np.percentile(exam_scores, percentile)
    plt.axvline(value, color='green', linestyle=':', 
                label=f'{percentile}th: {value:.1f}')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Professional-looking histogram with density curve
data = np.random.normal(100, 20, 1000)

plt.figure(figsize=(10, 6))

# Histogram with density curve
n, bins, patches = plt.hist(data, bins=25, density=True, alpha=0.7, 
                           color='lightseagreen', edgecolor='white', linewidth=0.5)

# Add a density curve
from scipy.stats import gaussian_kde
density = gaussian_kde(data)
xs = np.linspace(data.min(), data.max(), 200)
plt.plot(xs, density(xs), 'r-', linewidth=2, label='Density Curve')

# Customize the plot
plt.title('Professional Histogram with Density Curve', fontsize=14, fontweight='bold')
plt.xlabel('Values', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True, alpha=0.3)

# Add statistics box
stats_text = f'Mean: {data.mean():.2f}\nStd Dev: {data.std():.2f}\nN: {len(data)}'
plt.text(0.02, 0.98, stats_text, transform=plt.gca().transAxes, 
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.legend()
plt.show()
```

```python
# Comparing multiple groups
np.random.seed(42)
department_a = np.random.normal(75, 10, 100)
department_b = np.random.normal(82, 8, 100)
department_c = np.random.normal(68, 12, 100)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot([department_a, department_b, department_c], 
           labels=['Dept A', 'Dept B', 'Dept C'])
plt.title('Employee Performance by Department')
plt.ylabel('Performance Score')

# Enhanced box plot with individual points
plt.subplot(1, 2, 2)
data = [department_a, department_b, department_c]
positions = [1, 2, 3]

for i, (dept, pos) in enumerate(zip(data, positions)):
    # Box plot
    bp = plt.boxplot(dept, positions=[pos], widths=0.6, 
                    patch_artist=True, labels=[f'Dept {chr(65+i)}'])
    
    # Color the boxes
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
    for patch, color in zip(bp['boxes'], [colors[i]]):
        patch.set_facecolor(color)
        patch.set_alpha(0.6)
    
    # Add individual points with jitter
    jitter = np.random.normal(0, 0.05, len(dept))
    plt.scatter(np.full(len(dept), pos) + jitter, dept, 
               alpha=0.4, color=colors[i], s=30)

plt.title('Performance with Individual Points')
plt.ylabel('Performance Score')
plt.tight_layout()
plt.show()
```
